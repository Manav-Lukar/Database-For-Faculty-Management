```python
import pandas as pd

df = pd.read_excel('Journal Articles.xlsx')
df.head()
```

```python
df.drop(df.iloc[:, 5:], inplace=True, axis=1)
```

```python
def author_name(auth):
    for ind in df.index:
        if df['j_author'][ind].find(auth) != -1:
            print(df['j_author'][ind]," : " ,df['j_title'][ind])
```
